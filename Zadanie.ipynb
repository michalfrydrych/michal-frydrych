What countries receive the most medals nowadays? Does it change much across the
timespan?

```python
best_of =  data.Team.value_counts().sort_values(ascending = False).head(10)
best_of
```

```python
temp = data.sort_values('Year')
temp
```

```python
plt.figure(figsize = (12,8))
plt.title("Best_of")
sns.barplot(x = best_of.index , y = best_of);
```

Who wins the most medals? Women or men?

```python
Who_is_best = data.Medal.value_counts()
Who_is_best
```

```python
plt.figure(figsize = (15,12))
sns.countplot(x = 'Medal',data = data,hue = 'Sex');
```

What’s the trend for the number of participants?

```python
plt.figure(figsize = (12,8))
sns.countplot(x = 'Year',data = data)
plt.xticks(fontsize=12,rotation=90);
```

What is the relation between age and medals won?

```python
plt.figure(figsize = (20,20))
sns.boxplot(x="Medal", y="Age", data=data,palette='Set1')
```

Do tall people win more medals?

```python
plt.figure(figsize = (20,20))
sns.boxplot(x="Medal", y="Height", data=data,palette='Set1')
plt.xticks(fontsize=12,rotation=90);
```

Which sports are dominated by short people?

```python
plt.figure(figsize = (20,20))
sns.boxplot(x="Sport", y="Height", data=data,palette='Set1')
plt.xticks(fontsize=12,rotation=90);
```
